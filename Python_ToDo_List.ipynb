```python
#Todo using List
todo=[]
while True:
  print("--------------------")
  print("To Do List By Vishal")
  print("--------------------")
  print("1.Add Task\n2.Edit Task\n3.Delete Task\n4.List All Task\n0.Exit\n")
  choice=int(input("Enter:"))
  if choice==0: #Exit
    print("Exiting....bye")
    break
  elif choice==1: #Add Task
    task=input("Enter Task:")
    todo.append(task)
    print("List updated with task->",task)
  elif choice==2: #Edit
    print("------------------------")
    print("List has")
    print("------------------------")
    for index in range(len(todo)):
      print(index+1,")",todo[index])
    i = int(input("Enter task no to Update:"))
    if i-1<len(todo):   #size
      print("Old task:",todo[i-1])
      new_task=input("Enter new Task:")
      todo[i-1] = new_task
      print("List updated with task->",new_task)
    else:
      print("Task no out of list.")
  elif choice==3: #Delete
    print("------------------------")
    print("List has")
    print("------------------------")
    for index in range(len(todo)):
      print(index+1,")",todo[index])
    i = int(input("Enter task no to Delete:"))
    if i-1<len(todo):   #size
      deleted=todo.pop(i-1)
      print("Deleted:",deleted)
    else:
      print("Task no out of list.")
  elif choice==4: #List all
    print("------------------------")
    print("Your List has")
    print("------------------------")
    for index  in range(len(todo)):
      print(index+1,")",todo[index])

  else: #default for error handling
    print("Wrong option selected")


# print("Your choice is:",choice)
```
